```python
import pandas as pd
import seaborn as sns

df = pd.read_csv('dulieuxettuyendaihoc.csv',header=0, delimiter=',',encoding='utf-8')
```

```python
df.head(5)
```

```python

df = df[['T5','T6','GT','DT', 'KV', 'KT', 'NGONNGU','TOANLOGICPHANTICH','GIAIQUYETVANDE','NGAYTHI','DINHHUONGNGHENGHIEP']]
```

```python
#đổi tên cột
df.rename(columns={'TOANLOGICPHANTICH':'LOGIC',
                   'GIAIQUYETVANDE':'UNGXU',
                   'DINHHUONGNGHENGHIEP' : 'HUONGNGHIEP'},inplace=True)
df.head(5)
```

```python
#xóa bỏ các dòng dữ liệu rỗng
df.dropna(how='all', inplace=True)
```

```python
#xóa bỏ các dòng bị trùng
df.drop_duplicates(inplace=True)
df.head()
```

```python
#Xử lý dữ liệu thiếu
#đầu tiên trực quan xem dữ liệu bị thiếu
#xem có cần xử lý không
import matplotlib.pyplot as plt
```

```python
#DÙng hàm heatmap để trực quan hoá dữ liệu
plt.figure(figsize=(10,6))
sns.heatmap(df.isna().transpose(),cmap='YlGnBu',
            cbar_kws={'label':'Dữ liệu thiếu'})
plt.savefig('missingdata.png',dpi=100)
'''
note: với dữ liệu bị thiếu:
(1) Cần xác định biến số nào bị thiếu
(2) Mức độ thiếu dữ liệu
(3) Có cần phải xử lý không
'''
```

```python
#điền giá trị thiếu
#định tính : scalar, AI, Mode
#định lượng: Scalar, Mean, median, x/s,AI

df['DT'].fillna('KINH',inplace=True)
#Lưu ý với biến định tính ta có thể thay
#bằng giá trị yếu vị (mode)
#df['DT'].fillna(df['DT'].mode()[0],inplace=True)
```

```python
#điền thiếu giá trị phần NGONNGU bằng 0 nếu có
df['NGONNGU'].fillna(0,inplace=True)
#điền thiếu giá trị logic bằng trung bình nếu có
df['LOGIC'].fillna(df['LOGIC'].mean(),inplace=True)
#điền thiếu giá trị ứng xử bằng trung vị nếu có
df['UNGXU'].fillna(df['UNGXU'].median(),inplace=True)
#Lưu ý với biến định lượng thì ta nên thay bằng trung vị
#điền thiếu T6 bằng phương pháp interpolate ( nội quy)
#lưu ý: interpolate dùng cho dữ liệu mang tính thứ tự ( tuần tự)
```

```python
'''
Phần 2: tự tạo ra các biến số hay các cột dữ liệu mới
Nếu chỉ là xử lý phân tích dự liệu thì gọi là New Attribute
Đây là kỹ thuật tạo thêm hoặc biến đổi số liệu sẵn có thành các biến số mới phù hợp nghiệp vụ để phân tích 

Tạo biến toán TBTOAN: trung bình toán 12

'''
df['TBTOAN'] = (df['T5']+df['T6'])/2
```

```python
#tạo biến XEPLOAI: dựa trên df['TBTOAN]
df.loc[df['TBTOAN'] < 5.0, 'XEPLOAI'] = 'FAIL'

df.loc[(df['TBTOAN'] >= 5.0) & 
       (df['TBTOAN'] <= 7.0), 'XEPLOAI'] = 'FAIR'

df.loc[(df['TBTOAN'] >= 7.0) & 
       (df['TBTOAN'] <= 9.0), 'XEPLOAI'] = 'GOOD'

df.loc[(df['TBTOAN'] >= 9.0) , 'XEPLOAI'] = 'EXCEL'
```

```python
df.head(10)
```

```python
'''
Tạo biến khối thi thảo mãn ( g: group)
a1:g1
c:g3
d1:g3
a:g1
b:g2

filter -> map -> reduce forward
'''
dict_map = {
    'A1':'G1', 'C':'G3', 'D1':'G3', 'A':'G1','B':'G2'
}
df['NHOMKT'] = df['KT'].map(dict_map)
```

```python
df.head(5)
```

```python
'''
Tạo biến cộng: CONG

Nếu khối thi thuộc nhóm G1, G2 và TBTOAN >= 5.0 thì là CONG = 1.0
ngược lại thì là 0 ( ánh xạ NHOMKT==X,TBTOAN==Y)
'''
def fplus(x,y):
    if(x == 'G1' or x == 'G2') and (y >=5.0):
        return 1.0
    else:
        return 0.0
df['CONG'] = list(map(fplus,df['NHOMKT'],df['TBTOAN']))
```

```python
'''
ĐỘ XIÊN (skewness) cho biết sự nghiêng ngã của 1 phân phối

-Lệch trái: càng về bên trái thì xác suất càng tháp(mũi nhọn bên phải)
-Lệch phải: càng về bên phải thì xác suất càng tháp(mũi nhọn bên trái)

Diễn giải cho skewness

skewness > 0 -> mean > median : ta gọi là Positive Skewness hay lệch phải, 
tức là giá trị ngoại biên(outlier) nhận giá trị lớn sẽ đẩy trung bình về phía cuối

skewness < 0 -> mean < median : ta gọi là Negative Skewness hay lệch trái, 
tức là giá trị ngoại biên(outlier) nhận giá trị nhỏ sẽ đẩy trung bình về phía đầu

skewness = 0 -> mean = median = mode(yếu vị) : Phân phối không lệch còn được gọi là phân phối đều đối xứng
'''
df['NGONNGU'].skew()
```

```python
'''Kurtosis:Độ nhọn
Độ lớn càng gần 3 thì fit
Dưới 3 thì fat
Trên 3 thì thin

Theo pandas được tính theo Fisher's Kurtosis(đã chuẩn hóa giá trị về excess kurtosis theo mean = 0)
Thông thường để đánh giá hình dáng, độ nhọn ta dùng đại lượng axcess= kurtosis - 3:
>0 -> thin
=0 -> fit
<0 -> fat

Trị tuyệt đối excess kurtosis càng cao thì mức độ thin fat càng lớn (càng âm càng mập, càng dương càng ốm)
'''
#Biến NGONNGU có độ nhọn ntn?? -> FAT
#Với phân phối chuẫn excess kurtosiss = 0, skewness = 0, kurtosis = 3
df[['NGONNGU']].kurtosis()
```

```python
'''
Phân tích sự tác động(ảnh hưởng) qua lại giữa 2 biến định lượng

Phương pháp 1: Hiệp phương sai:co-variance
-   Giá trị co-variance > 0 thì 2 biến có tương quan thuận(đồng biến)
-   Giá trị co-variance < 0 thì 2 biến có tương quan nghịch(nghịch biến)
-   Độ lớn(trị tuyệt đối giá trị) càng lớn thì mức độ quan hệ tương quan càng chặt chẽ

Ma trận hiệp phương sai co-variance matrix
'''
df[['T5','T6']].cov()
```

```python
#so sánh mức độ tương quan giữa T5 so với Logic và T6 so với logic(điểm toán trung bình của học kỳ nào có ảnh hưởng đến điểm LoGic)
'''
T5 so với logic = 0.289
T6 so với logic = 0.4 lớn hơn nên T6 ảnh hưởng đến Logic nhiều hơn
'''
df[['T5','T6','LOGIC']].cov()
```

```python
'''
Với phương pháp so sánh tương quan bằng covariance thì ta không đo lường được cường độ tương quan giữa 2 biến định lượng
Giải pháp: Tương quan tuyến tính Pearson
r nằm trong khoảng [-1;1]
r = 0 => không tương quan
r < 0: tương quan nghịch
r > 0:tương quan thuận
|r| càng gần 1 thì tương quan càng cao
|r| <0.5: thì tương quan thấp
[0.5 ; 0.65]: khá
[0.65 ; 0.75]: tốt
[0.75 ; 0.9]: rất tốt
> 0.9: hoàn hảo
'''
#khảo sát sự tương giữa T5 và T6 là 0.77-> rất tốt , >0 -> tương quan thuận
# nếu 2 cái này tương quan càng cao thì các điểm trên biểu đồ càng nằm sát trên đường thẳng
df[['T5','T6']].corr()
```

```python
'''
Trực quan hóa tương quan tuyến tính giữa 2 biến định lượng 
khám phá tương quan tuyên tính của 2 biến định lượng
thông qua biểu đồ phân tán (Scatter)
'''
sns.lmplot(data=df,x='T5',y='T6',fit_reg=True)
plt.show()
```
